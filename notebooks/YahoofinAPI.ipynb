```python
import time

def get_data_with_throttle(ticker, start_date, end_date, index_as_date=True, interval="1wk"):
    url = f"https://query1.finance.yahoo.com/v8/finance/chart/{ticker}?interval={interval}"
    response = requests.get(url)
    
    if response.status_code == 429:
        print("Rate limit hit. Waiting before retrying...")
        time.sleep(120)  # Wait for 10 seconds (or longer if needed)
        return get_data_with_throttle(ticker, start_date, end_date, index_as_date, interval)
    
    response.raise_for_status()  # Raise an error for other HTTP issues
    return response.json()

# Example usage
data = get_data_with_throttle("amzn", "12/04/2009", "12/04/2019")
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://www.londonstockexchange.com/exchange/prices-and-markets/stocks/indices/ftse-all-share.html"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Find tickers from the web page
tickers = []
for item in soup.find_all('a', href=True):
    if 'equity' in item['href']:
        tickers.append(item['href'].split('/')[-1])  # Extract ticker from URL

print("All UK Stock Tickers:", tickers)
```
